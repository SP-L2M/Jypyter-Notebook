```
import pandas as pd
import numpy as np
import plotly.express as px
```

```

url = 'https://raw.githubusercontent.com/SP-L2M/Jypyter-Notebook/master/sbi-data.csv'
df = pd.read_csv(url)
time_diff = np.diff(df["Time"])
fig = px.line(time_diff)

df = pd.DataFrame(time_diff, columns = ['Time']) 
fig = px.line(df, y="Time", title='SBI Timestamp plot (Sample Number Vs Sample Time Interval (micro-seconds))', height=1000)

fig.show()
```
